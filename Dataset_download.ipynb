```python
def print_number(split):
  """
  Prints out the number of images and labels in the specified dataset split
  """

  imgs = len(glob.glob(f"/content/drive/MyDrive/Licence Plate OCR/dataset/{split}/images/*.jpg"))
  lbs =  len(glob.glob(f"/content/drive/MyDrive/Licence Plate OCR/dataset/{split}/labels/*.txt"))
  return f"Number of {split} images: {imgs} \n Number of labels: {lbs}"
```
